### Convert raw data(csv file) to lists
- Used to process csv data by converting data into list type
- csv_to_lists(f_path, f_name)
  - f_path: e.g) "./"
  - f_name: e.g) "lt000.csv"
  - return: 'temp_line_data': list to be converted from 'lt000.csv'

```python
def csv_to_lists(f_path, f_name):
    lt_file = open(f'{f_path}{f_name}','r')
    data_read = csv.reader(lt_file)
    
    data_fcnt = 0
    data_lcnt = 0
    temp_line_data = []
    
    for line_data in data_read:
        # print(line_data)
        temp_line_data.append(line_data)
        data_fcnt = data_fcnt + 1
    # for i in range(10):
    #     print(all_line_data[i])
    
    lt_file.close()
    return temp_line_data
```

- Lists that have all the data from csv file
  - 'all_line_data' to be the fundamental data that are made use of

```python
all_line_data = csv_to_lists("./", "lt000.csv")
```

### Save list to csv file
- Used to create datasets for training model
- save_to_csv(data, slice_low, slice_high, idx, f_name, f_path)
  - Definition: save list to csv files by slicing with the number of slice_idx
  - data: data type like 'all_line_data' is used, and possible to slice
  - slice_low: starting point of slicing(from LSB)
  - slice_high: ending point + one digit of slicing(from MSB)
  - idx: used as index of csv file that is generated, for numbering file name
  - f_path: file directory to save files generated. e.g) "./data_ready/1/"

```python
def save_to_csv(data, slice_low, slice_high, idx, f_name, f_path):

    file_path = f_path + f'{f_name}{idx:04d}.csv'

    # Convert data to strings if they are not already strings
    data = [[str(value) for value in row[slice_low:slice_high]] for row in data]

    with open(file_path, 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerows(data)

    print(f'The data has been saved to {file_path}')
```

### Data grouping with specified range
- group_range_csv(data, start_index, group_size)
  - Definition: grouping data by num_group
  - data: data type like 'all_line_data' is used
  - group_size: number of list in one group. e.g) 10 lists in a row are included from 'data'(all_line_data) in case 'group_size' = 10
  - file_number: For numbering file name

```python
def group_range_csv(data, start_index, group_size, file_number, file_path):
    save_to_csv(data[start_index: start_index+group_size], file_number, file_path)
```
